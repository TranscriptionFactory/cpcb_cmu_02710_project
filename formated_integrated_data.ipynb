```python
import pandas as pd
import os
```

```python
# expression matrix
exp_path = 'TcellDev/Data/T_cell_development/integrated/t_data.txt'

# cluster ids and cell type ids
id_path = 'TcellDev/Data/T_cell_development/integrated/cluster_pseudotime.txt'
```

```python
exp = pd.read_csv(exp_path, sep = "\t")
```

```python
exp = exp.rename(columns = {'Unnamed: 0':'gene_id'})
```

```python
exp.head()
```

```python
exp.shape
```

```python
id_info = pd.read_csv(id_path, sep = "\t")
id_info = id_info.rename(columns = {'Unnamed: 0':'barcode'})
```

```python
id_info
```

```python
# transpose 
exp_alt = exp.transpose()
```

```python
exp_alt.columns = exp_alt.loc['gene_id']

exp_alt = exp_alt.drop(labels = 'gene_id', axis = 0)
```

```python
exp_alt['barcode'] = exp_alt.index
```

```python
exp_alt = exp_alt.reset_index(drop = True)
```

```python
exp_alt
```

```python
# left join to get IDs for barcodes
exp_df_join = exp_alt.merge(id_info[['barcode', 'bigstage']], left_on = 'barcode',
                          right_on = 'barcode')
```

```python
exp_df_join
```

```python
# rename these to match
exp_df_join['bigstage'] = exp_df_join['bigstage'].replace({'DN': '1DN', 'ISP': '2preDP', 'DPbla':'3DPbla',
                                        'DPre': '4DPre', 'DPsel': '5DPsel', 'CD4': '6CD4sp',
                                        'CD8': '7CD8sp'})


# array(['6CD4sp', '10APC', '5DPsel', '4DPre', '2preDP', '3DPbla', '1DN',
#        '7CD8sp', '9NCL']
```

```python
# add time column
exp_df_join['time'] = exp_df_join['bigstage'].replace({'1DN':1, '2preDP':2, '3DPbla':3, '4DPre':4,
                                          '5DPsel':5, '6CD4sp':6, '7CD8sp':6})
```

```python
exp_df_join
```

```python
dfcols = exp_df_join.columns.tolist()
reorder_cols = dfcols[-3:] + dfcols[:-3]
```

```python
exp_df2 = exp_df_join[reorder_cols]
```

```python
# put time before bistage
dfcols = exp_df2.columns.tolist()
reorder_cols = [dfcols[0], dfcols[2], dfcols[1]] +  dfcols[3:]
```

```python
exp_df2 = exp_df2[reorder_cols]
```

```python
# drop cols with all zeros
exp_df3 = exp_df2.loc[:, (exp_df2 != 0).any(axis = 0)]
```

```python
exp_df3
```

```python
# save
exp_df3.to_csv('integrated_data.E', sep = "\t", index = False)
```
