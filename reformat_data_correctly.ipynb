```python
import pandas as pd
import os
```

```python
# expression matrix
exp_path = 'TcellDev/Data/T_cell_development/10x_1993_log2_q_norm.txt'

# cluster ids and cell type ids
id_path = 'TcellDev/Data/T_cell_development/cell_cluster15_self_modcell.txt'
```

```python
exp = pd.read_csv(exp_path, sep = "\t")
```

```python
exp.head()
```

```python
exp.shape
```

```python
id_info = pd.read_csv(id_path, sep = "\t")
```

```python
id_info
```

```python
# pivot column names into a row
barcodes_exp = exp.columns[1:]
```

```python
barcodes_exp
```

```python
# pivot to get column names (barcodes) into rows in list
exp_df = exp.melt(id_vars = ['gene_id'],value_vars = barcodes_exp,
                 var_name = 'barcode', value_name = 'count')
```

```python
exp_df
```

```python
# left join to get IDs for barcodes
exp_df_join = exp_df.merge(id_info[['cell', 'bigstage']], left_on = 'barcode',
                          right_on = 'cell')
```

```python
exp_df_join
```

```python
# slightly easier way to make this df
exp_alt = exp.transpose()
```

```python
exp_alt.columns = exp_alt.loc['gene_id']

exp_alt = exp_alt.drop(labels = 'gene_id', axis = 0)
```

```python
# make column for barcode
gene_names= exp_alt.columns[1:]

exp_alt['barcode'] = exp_alt.index

# exp_alt = exp_alt.reset_index()

exp_df2 = exp_alt.merge(id_info[['cell', 'bigstage']], left_on = 'barcode',
                          right_on = 'cell')
# exp_join = exp_alt.merge(id_info[['cell', 'bigstage']], left_on = 'barcode',
#                           right_on = 'cell')
```

```python
exp_df2 = exp_df2.drop(labels = 'cell', axis = 1)
exp_df2

# exp_df2 = exp_df2[['barcode', 'bigstage', gene_names]]
```

```python
dfcols = exp_df2.columns.tolist()
reorder_cols = dfcols[-2:] + dfcols[:-2]
```

```python
exp_df3 = exp_df2[reorder_cols]
```

```python
# exp_df3.to_csv('pivoted_exp.csv', sep = "\t")
```

```python
exp_df3
```

```python
# unique values
exp_df3['bigstage'].unique()
```

```python
tcell_stages = id_info['bigstage'].unique()[:-2]
```

```python
exp_df4 = exp_df3.loc[exp_df3['bigstage'].isin(tcell_stages)]
```

```python
exp_df4['bigstage'].unique()
```

```python
#rename
exp_df5 = exp_df4.copy()
# exp_df5 = exp_df4.rename(columns = {"barcode": "index", "bigstage": "time", ""})
```

```python
exp_df5['time'] = exp_df5['bigstage']
```

```python
exp_df5['time'] = exp_df5['time'].replace({'1DN':1, '2preDP':2, '3DPbla':3, '4DPre':4,
                                          '5DPsel':5, '6CD4sp':6, '7CD8sp':6})
```

```python
cols_rename = exp_df5.columns.tolist()
```

```python
cols_reorder = [cols_rename[0]] + [cols_rename[-1]] + cols_rename[1:-1]
```

```python
exp_df5 = exp_df5[cols_reorder]
```

```python
exp_df5 = exp_df5.rename(columns = {'barcode':'cell_name', 'time': 'time_point', 'bigstage':'assignment'})
exp_df5 = exp_df5.reset_index(drop = True)
exp_df5
```

```python
exp_df5.to_csv('tcell_data2.E', sep = "\t", index = False, index_label = False)
```
